```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = np.load("Covid_Decided_Count.npy")
data
```

```python
first_digit = [int(x[0]) for x in data.astype("str")]
cnt = np.bincount(first_digit)[1:]
cnt
```

```python
plt.bar(range(1, 10), cnt, color="lawngreen")
plt.xticks(range(1, 10))
plt.show()
```

```python
plt.pie(cnt, labels=list(map(str, range(1, 10))), autopct="%.2f")
plt.show()
```
